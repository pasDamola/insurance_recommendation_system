```python
df = pd.read_csv('Train.csv')
df.head()
```

```python
#Find out the age of each customer as at 2020
df['age'] = df.birth_year.apply(lambda x: 2020 - int(x))
df
```

```python
#Seperate necessary features for predicition
#We are ruling out the ID, join_date and branch_code because from intuition they may not be useful for prediction
feature_columns = df[['sex', 'marital_status', 'age', 'occupation_code', 'occupation_category_code']]
feature_columns
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

feature_labelled_sex = le.fit_transform(feature_columns.sex)
feature_labelled_marital_status = le.fit_transform(feature_columns.marital_status)
feature_labelled_occupation_category_code = le.fit_transform(feature_columns.occupation_category_code)
```

```python
feature_columns['sex'] = feature_labelled_sex
feature_columns['marital_status'] = feature_labelled_sex
feature_columns['occupation_category_code'] = feature_labelled_sex
feature_columns
```
